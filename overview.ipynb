```python
import altair as alt
import pandas as pd
import panel as pn
import numpy as np
from itertools import product
alt.data_transformers.disable_max_rows();
pn.extension('vega')
```

```python
csv = pd.read_csv("overview.csv",index_col=0,header=0,sep=",",engine='python');
data = pd.DataFrame(list(product(csv.columns.values,csv.index.values)), columns=['Year', 'Country']);
data['Annual Anomaly'] = csv.values.transpose().ravel();

with open('continents.txt') as f:
        continents = f.read().splitlines()

data['Continent'] = continents
    
def overview():
    return pn.Row(alt.Chart(data).mark_rect().encode(
            x = alt.X('Year:O',title = 'Year'),
            y = alt.Y('Country:O',title = None, axis = alt.Axis(labelAngle = -7)),
            color = alt.Color('Annual Anomaly:Q', scale = alt.Scale(scheme = "redblue", reverse = True, domainMin = -2, domainMid = 0, domainMax = 2.5)),
            tooltip = ['Country','Year','Annual Anomaly','Continent'],
        ).properties(
            height = 500,
            width = 1000
       # ).configure_legend(
        #    titleOrient = 'right',
         #   gradientLength = 400,
          #  gradientThickness = 30
        ).facet(
            row=alt.Row('Continent:O', header=alt.Header(labelFontSize=18),title=None, sort = ['Africa','Asia','Europe','North America','South America','Oceania',''])
        ).resolve_scale(
          y='independent'
        ))
html = '''
    <h1>Visualisation of Climate Change</h1>
    <div class="row-fluid">
        <div style="margin-right" class="span4 text-left"><form action='https://berkeleydatavis.herokuapp.com/start'><input type='submit' value='Take me back to the Start' style='float: right'/></form></div>
        <div style="" class="span4 text-center"><form action='https://berkeleydatavis.herokuapp.com/detail'><input type='submit' value='Detailed View' style='float: right'/></form></div>
        <div style=""class="span4 text-right"><form target='_blank' action='https://berkeleydatavis.herokuapp.com/faq'><input type='submit' value='FAQ' style='float: right'/></form></div>
    </div>
    '''
view = pn.Column(pn.pane.HTML(html),overview)

view.servable()
```
