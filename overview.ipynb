```python
import altair as alt
import pandas as pd
import numpy as np
from itertools import product
alt.data_transformers.disable_max_rows();
```

```python
data = pd.read_csv("overview.csv",index_col=0,header=0,sep=",",engine='python');
test = pd.DataFrame(list(product(data.columns.values,data.index.values)), columns=['x', 'y']);
test['z'] = data.values.transpose().ravel();

alt.Chart(test).mark_rect().encode(
    x = alt.X('x:O',title = 'Year'),
    y = alt.Y('y:O',title = 'Country', axis = alt.Axis(labelAngle = -7)),
    color = alt.Color('z:Q', scale = alt.Scale(scheme = "redblue", reverse = True, domainMid = 0)),
    tooltip = ['y:O','x:O','z:Q']
).properties(
    height = 2000
)
```
