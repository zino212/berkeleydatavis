```python
import altair as alt
import pandas as pd
import panel as pn
from itertools import product
alt.data_transformers.disable_max_rows();
pn.extension('vega')
```

```python
data = pd.read_csv("overview.csv",index_col=0,header=0,sep=",",engine='python');
test = pd.DataFrame(list(product(data.columns.values,data.index.values)), columns=['Year', 'Country']);
test['Annual Anomaly'] = data.values.transpose().ravel();

def overview():
    return pn.Row(alt.Chart(test).mark_rect().encode(
            x = alt.X('Year:O',title = 'Year'),
            y = alt.Y('Country:O',title = 'Country', axis = alt.Axis(labelAngle = -7)),
            color = alt.Color('Annual Anomaly:Q', scale = alt.Scale(scheme = "redblue", reverse = True, domainMin = -2, domainMid = 0, domainMax = 2.5)),
            tooltip = ['Country','Year','Annual Anomaly']
        ).properties(
            title = 'Overview',
            height = 2500
        ).configure_legend(
            titleOrient = 'right',
            gradientLength = 400,
            gradientThickness = 30
        ))

view = pn.Column(pn.pane.HTML("<h1>Visualisation of Climate Change</h1><form action='https://berkeleydatavis.herokuapp.com/detail'><input type='submit' value='Detailed View'/></form><br><form action='https://berkeleydatavis.herokuapp.com/faq'><input type='submit' value='FAQ'/></form>"),overview)

view.servable()
```
